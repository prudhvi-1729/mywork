```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
class Linkedlist():
    def __init__(self):
        self.head=None
    def insertFront(self,data):
        node=Node(data,self.head)
        self.head=node
    def insertLast(self,data):
        newNode=Node(data)
        k=self.head
        while(k.next!=None):
            k=k.next
        k.next=newNode
    def getLength(self):
        c=0
        p=self.head
        while p:
            c=c+1
            p=p.next
        return c    
    def insertAt(self,pos,data):
        if(pos<=0 or pos>self.getLength()):
            raise Exception("invalid index")
        if pos==0:
            self.insertFront(data)
            return
        count=0
        itr=self.head
        while itr:
            if(count==pos-1):
                new=Node(data,itr.next)
                itr.next=new
                break
            itr=itr.next
            count=count+1
    def removeAt(self,pos):
        if(pos<=0 or pos>self.getLength()):
            raise Exception("invalid index")
        if pos==0:
            self.head=self.head.next
            return
        itr=self.head
        count=0
        while itr:
            if(count==pos-1):
                itr.next=itr.next.next
                break
            itr=itr.next
            count=count+1
    def insertAfterValue(self,val,data):
        current=self.head
        while current!=None:
            if(current.data==val):
                newNode=Node(data,current.next)
                current.next=newNode
            current=current.next
    def removeAfterValue(self,val):
        current=self.head
        while current!=None:
            if(current.data==val):
                current.next=current.next.next
            current=current.next        
    def display(self):
        current=self.head
        while current!=None:
            print(current.data,end=" ")
            current=current.next
            
k=Linkedlist()
k.insertFront(14)
k.insertFront(13)
k.insertLast(15)
k.insertLast(16)
k.insertFront(12)
k.insertFront(10)
k.insertLast(17)
k.insertAt(3,100)
k.removeAt(3)
k.insertAfterValue(10,11)
k.removeAfterValue(13)
k.display()        
```
